```python
import numpy as np
import pandas as pd

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func


#################################################
# Database Setup
#################################################
engine = create_engine("postgresql://postgres:admin@localhost/fultoncounty_db")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
df1 = pd.read_sql_table('home_sales', engine)
df1.head()
```

```python
engine2 = create_engine('sqlite:///home_sales.sqlite')

with engine2.connect() as connection:
    df1.to_sql(name='home_sales', con=connection, index=False, if_exists='replace')
```
